### Rules of Matrix Multiplication
There are two main rules:
1. Inner dimensions must match
2. Resulting vector should have shape of **outer dimensions**

*    (3,2) * (3,2)   won't work
*    (3,2) * (2,3)   will work
*    (2,3) * (3,2)   will work





```python
torch.matmul(torch.rand(2,10), torch.rand(10,2))
```

### Shapes for matrix multiplication

```python
# Shapes

tensor_A = torch.tensor([[1,2],
                          [3,4],
                          [5,6]])

tensor_B = torch.tensor([[2,3],
                        [6,0],
                         [9,7]])

torch.matmul(tensor_A,(tensor_B.T))
# torch.matmul(tensor_A,tensor_B) -> gives error due to incorrect shape
```

### Tensor Aggregation

Min, Max, Mean, Sum etc(aggregation)

```python
# Find Minimim

x = torch.arange(0,100,10)
min = torch.min(x), x.min()
min, x.dtype
```

```python
# Find Maximum

max = torch.max(x), x.max()
max
```

```python
# Find mean

mean = torch.mean(x.type(dtype=torch.float32)), x.type(torch.float32).mean()
mean
```

```python
# Find sum

sum = torch.sum(x), x.sum()
sum
```

```python
# Find argmin

argmin = torch.argmin(x), x.argmin()
argmin, x[0]
```

```python
# Find argmax

argmax = torch.argmax(x), x.argmax()
argmax, x[9]
```

### Reshaping, Stacking, Squeezing and Unsqueezing tensors

```python
# Create a tensor

y = torch.arange(1.0,11.0)
y, y.shape
```

```python
# Reshape

y_reshaped =y.reshape(1,10)
y_reshaped, y_reshaped.shape
```

```python
# Change the view
# view returns the input tensor dimensions but shares the same memory
# as original input, also changing view will change the original input

z = y.view(1,10)
z, z.shape
```

```python
z[:,0] = 5
z,y
```

```python
torch.vstack([y,y,y,y])
```

```python
torch.hstack([y,y,y,y])
```

```python
# Squeeze -> removes all the 1 dimensions

squeezed = torch.squeeze(y_reshaped, dim=0)
print(f"Original tensor: {y_reshaped}")
print(f"Original tensor shape: {y_reshaped.shape}")
print(f"\nSqueezed tensor: {squeezed}")
print(f"Squeezed tensor shape: {squeezed.shape}")
```

```python
# Unsqueeze -> adds the 1 dimension

unsqueezed = torch.unsqueeze(squeezed, dim=0)

print(f"Original tensor: {y}")
print(f"Original tensor shape: {y.shape}")
print(f"\nUnsqueezed tensor: {unsqueezed}")
print(f"Unsqueezed tensor shape: {unsqueezed.shape}")
```

```python
# Permute -> rearranges the dimensions of a target tensor in a specified order
# permute is a view

x_original = torch.rand(size=(224,224,3))
x_original[0,0,0] = 876
x_permuted = x_original.permute(2,0,1)
x_permuted.shape, x_permuted[0,0,0]


```
